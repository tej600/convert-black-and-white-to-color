```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()

# Get image name
image_name = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_name)

# Check image
if img is None:
    print(" Error loading image")
    exit()

# Convert to RGB (for display)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply threshold
_, mask = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

# Copy image
colorized = img.copy()

# Apply green color where mask is white
colorized[mask == 255] = [0, 255, 0]

# Show images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Colorized Image")
plt.imshow(cv2.cvtColor(colorized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()

# Save output
cv2.imwrite("output.jpg", colorized)

print(" Saved as output.jpg")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
image_name = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_name)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Copy image
colorized = img.copy()

# Apply MULTI-LEVEL COLORS (based on intensity)

# Dark pixels → Blue
colorized[gray < 80] = [255, 0, 0]

# Medium-dark → Purple
colorized[(gray >= 80) & (gray < 150)] = [255, 0, 255]

# Medium-bright → Yellow
colorized[(gray >= 150) & (gray < 200)] = [0, 255, 255]

# Bright → Green
colorized[gray >= 200] = [0, 255, 0]

# Show images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Multi-Color Output (4 Colors)")
plt.imshow(cv2.cvtColor(colorized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()

# Save & download
cv2.imwrite("output.jpg", colorized)
files.download("output.jpg")

print("Saved and downloaded as output.jpg")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
image_name = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_name)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply MULTI-COLOR using colormap (many colors)
colorized = cv2.applyColorMap(gray, cv2.COLORMAP_TURBO)

# Show images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Colorized Image (Many Colors)")
plt.imshow(cv2.cvtColor(colorized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()

# Save & download
cv2.imwrite("output.jpg", colorized)
files.download("output.jpg")

print("Saved and downloaded as output.jpg")
```
